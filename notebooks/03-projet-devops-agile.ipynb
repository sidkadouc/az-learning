## 1. Le cycle de vie d'un projet IT

Tout projet logiciel — qu'il déploie une appli web, un agent IA, ou un dataset — passe par les mêmes grandes étapes :

```
  ┌────────────┐   ┌────────────┐   ┌──────────┐   ┌──────────┐   ┌─────────┐   ┌──────────┐
  │  Cadrage   │→  │ Conception │→  │ Réalisat.│→  │  Tests   │→  │ Déploi. │→  │   Run    │
  │ (besoin,   │   │ (archi,    │   │  (code)  │   │ (unit,   │   │ (prod)  │   │ (monito- │
  │  user      │   │  design,   │   │          │   │  intégr.,│   │         │   │  ring,   │
  │  stories)  │   │  specs)    │   │          │   │  e2e)    │   │         │   │  support)│
  └────────────┘   └────────────┘   └──────────┘   └──────────┘   └─────────┘   └──────────┘
```

La grande question : **on fait ces étapes en cascade (Waterfall) ou en boucle (Agile) ?**

## 2. Waterfall vs Agile

### Waterfall (cascade) — l'ancien monde

Toutes les étapes ci-dessus, **dans l'ordre, en bloc, sur des mois**. On figeait les specs, puis on codait, puis on testait, puis on livrait.

👎 Problèmes : besoin client qui change en cours de route, gros effet tunnel, on découvre les bugs trop tard, on livre 12 mois plus tard un produit déjà obsolète.

### Agile — l'approche moderne

On livre **par petites itérations** (1 à 4 semaines). À chaque cycle on cadre + code + teste + livre **une petite tranche** de valeur. Le besoin peut évoluer entre deux cycles.

```
    Waterfall                              Agile
    ─────────                              ─────
                                           ┌──────────────────┐
    [Specs] ────────►                      │ Sprint 1 (2 sem) │ → livraison v0.1
    [Design] ────────►                     ├──────────────────┤
    [Code] ──────────────►                 │ Sprint 2 (2 sem) │ → livraison v0.2
    [Tests] ──────────────►                ├──────────────────┤
    [Deploy] ──────────────────►           │ Sprint 3 (2 sem) │ → livraison v0.3
                                           └──────────────────┘
    🕐 12 mois pour livrer                  🚀 valeur dès 2 semaines
    ❌ pas d'ajustement                     ✅ feedback à chaque cycle
```

Le **Manifeste Agile** (2001) tient en 4 valeurs :

- **Les individus et leurs interactions** plus que les processus et les outils
- **Des logiciels opérationnels** plus que la documentation exhaustive
- **La collaboration avec les clients** plus que la négociation contractuelle
- **L'adaptation au changement** plus que le suivi d'un plan

📚 https://agilemanifesto.org/iso/fr/

## 3. Scrum en pratique

**Scrum** est le framework agile le plus utilisé. Il définit des **rôles**, des **artefacts** et des **rituels** autour d'itérations de longueur fixe (typiquement **2 semaines**) appelées **sprints**.

### 3.1 Les rôles

| Rôle | Ce qu'il fait | À qui il parle |
|------|---------------|----------------|
| **Product Owner** (PO) | Décide **quoi** construire et dans quel ordre. Gère le backlog. | Métier, utilisateurs, équipe |
| **Scrum Master** (SM) | Facilite le process, débloque l'équipe, anime les rituels. | L'équipe, le PO, le management |
| **Équipe Dev** | Réalise (dev, design, test, déploiement). 3 à 9 personnes en général. | Entre eux + PO + SM |
| **Stakeholders** | Sponsors, métier, utilisateurs finaux. Donnent du feedback aux demos. | PO surtout |

### 3.2 Les artefacts

- **Product Backlog** : la grande liste des choses à faire (priorisée par le PO).
- **Sprint Backlog** : ce qu'on s'engage à faire **dans ce sprint**.
- **Incrément** : ce qui a été livré à la fin du sprint (idéalement déployable).
- **User Story** : un élément du backlog, sous la forme :  
  *« En tant que &lt;rôle&gt;, je veux &lt;action&gt; afin de &lt;bénéfice&gt;. »*  
  Ex. *« En tant que stagiaire, je veux déployer une web app pour comprendre App Service. »*

### 3.3 Les cérémonies (rituels)

```
                ┌─────────────────────────── 2 semaines ──────────────────────────┐
                ▼                                                                  ▼
                ┌───────────┐ ┌─────────────────────────────────────┐ ┌──────────┐ ┌──────────┐
                │ Sprint    │ │  Daily (15 min, debout, tous les jours) │ │  Sprint  │ │  Sprint  │
                │ planning  │ │                                          │ │  review  │ │  retro   │
                │ (1-2 h)   │ │  - Hier j'ai fait quoi ?                 │ │  (demo)  │ │  (qu'est-│
                │           │ │  - Aujourd'hui je fais quoi ?            │ │          │ │ ce qu'on │
                │ Qu'est-ce │ │  - Bloqueurs ?                           │ │ On montre│ │ améliore?│
                │ qu'on     │ └─────────────────────────────────────┘ │ aux stake-│ └──────────┘
                │ prend ?   │       (+ backlog refinement en cours       │ holders   │
                └───────────┘        de sprint, 1h/sem)                  └──────────┘
```

| Cérémonie | Durée | But |
|-----------|-------|-----|
| **Sprint Planning** | 1-2 h | Choisir les user stories du sprint |
| **Daily standup** | 15 min | Synchro rapide debout (hier / aujourd'hui / bloqueurs) |
| **Backlog Refinement** | 1 h / sem | Préciser et estimer les prochaines stories |
| **Sprint Review** | 1 h | Démo aux stakeholders, feedback |
| **Sprint Retrospective** | 1 h | L'équipe parle d'elle-même : qu'a-t-on appris, qu'améliorer ? |

> 🃏 **Estimation en story points** (1, 2, 3, 5, 8, 13… Fibonacci) ou en **t-shirt sizes** (S / M / L). On utilise souvent un **planning poker** : chacun montre sa carte, on discute des écarts. C'est une **estimation relative**, pas un engagement précis en heures.

## 4. 🎬 Mise en situation 1 — Une nouvelle feature arrive

**Contexte :** vous êtes dans une équipe Scrum de 5 personnes, sprint de 2 semaines. Le PO arrive lundi avec une demande métier.

**Jour 1 (lundi) — Cadrage rapide**
- 📝 Le PO discute avec le métier : *« On veut que les utilisateurs puissent télécharger leur facture en PDF. »*
- Le PO crée une **user story** dans Jira/Azure Boards.

**Jour 2 (mardi) — Refinement**
- 🧑‍💻 L'équipe et le PO se réunissent 30 min : on précise les critères d'acceptation (« le PDF doit contenir le logo, le total TTC, la date… »).
- On découpe en **3 sous-tâches** : (1) endpoint API qui génère le PDF, (2) bouton frontend, (3) tests E2E.
- Estimation : **5 points**.

**Jour 3 → 9 (mercredi → mercredi suivant) — Réalisation**
- 👩‍💻 Un dev backend prend la tâche 1, fait une **branche Git** `feature/download-invoice-pdf`.
- 👨‍💻 Un dev frontend prend la 2.
- Chacun ouvre une **Pull Request** quand c'est prêt. Les autres relisent. CI verte → merge.
- 🧪 Les tests E2E (tâche 3) tournent en pipeline.

**Jour 10 (jeudi de la semaine 2) — Sprint Review**
- 🎬 Démo aux stakeholders : *« Voici comment télécharger sa facture. »*
- Feedback : *« Et en plusieurs langues ? »* → nouvelle story pour le backlog.

**Jour 10 (après la review) — Retro**
- 💬 L'équipe : *« On a sous-estimé les CSS du PDF, +1 j de retard. »* → décision : pour les prochains PDF, prévoir une réserve de temps.

**Lundi suivant — nouveau Sprint Planning.** Le cycle repart.

## 5. DevOps — le mur entre Dev et Ops, et comment le démolir

Historiquement : les **Devs** écrivent du code, le « jettent par-dessus le mur » à l'équipe **Ops** qui se débrouille pour le déployer et le faire tourner en prod. Résultat : finger-pointing, lenteur, incidents.

**DevOps** = une culture + des outils pour **fusionner** Dev et Ops, automatiser le maximum et raccourcir le délai entre *« commit »* et *« en prod »*.

### Les 3 piliers DevOps

| Pilier | Idée | Outil typique |
|--------|------|---------------|
| **Automatisation** | Tout ce qui est répétitif doit être un script ou un pipeline. | GitHub Actions, Azure Pipelines |
| **Mesure (observabilité)** | On ne peut pas améliorer ce qu'on ne mesure pas. | Application Insights (cf. notebook 06), Grafana |
| **Partage / Culture** | Les équipes partagent la responsabilité du run. | Wiki, post-mortems sans blâme |

### Le boucle DevOps (∞)

```
                    ┌────► PLAN ────► CODE ────► BUILD ────► TEST ──┐
                    │                                                │
                    │                                                ▼
                    └◄ MONITOR ◄ OPERATE ◄ DEPLOY ◄ RELEASE ◄────────┘
                                                      
        ◄────── Dev ──────► ◄────── Ops ──────►
                       (frontière qui s'estompe)
```

## 6. CI/CD — l'automatisation du chemin code → prod

- **CI** *Continuous Integration* : à chaque push sur Git, on **build** + on **teste** automatiquement. Si ça casse, on le sait dans la minute.
- **CD** *Continuous Delivery* : à chaque merge sur `main`, on déploie automatiquement (en staging, puis en prod après approbation).
- **CD** *Continuous Deployment* (variante) : on déploie en prod **sans approbation manuelle**. Réservé aux équipes très matures.

### Anatomie d'un pipeline CI/CD

```
  push sur Git
       │
       ▼
 ┌────────────┐    ┌────────────┐    ┌────────────┐    ┌────────────┐    ┌────────────┐
 │  CHECKOUT  │───►│   LINT     │───►│   BUILD    │───►│   TESTS    │───►│  PACKAGE   │
 │  (clone le│    │ (style code│    │ (compile,  │    │ (unit,     │    │ (Docker /  │
 │   repo)   │    │  ✗ stops)  │    │  bundle)   │    │  integ.)   │    │  zip)      │
 └────────────┘    └────────────┘    └────────────┘    └────────────┘    └────────────┘
                                                                              │
                                                                              ▼
 ┌────────────────────────────────────────────────────────────────────────────────────┐
 │                              ÉTAPES DE DÉPLOIEMENT                                 │
 │                                                                                    │
 │   ┌──────────────┐         ┌──────────────┐         ┌──────────────┐               │
 │   │   DEV env    │  ──►    │ STAGING env  │  ──►    │   PROD env   │               │
 │   │ (auto)       │         │ (auto)       │         │ (approbation)│               │
 │   └──────────────┘         └──────────────┘         └──────────────┘               │
 └────────────────────────────────────────────────────────────────────────────────────┘
```

Si **une étape échoue**, le pipeline s'arrête. Le commit est marqué ❌ dans Git. Le dev répare puis re-push.

### Outils chez Microsoft

| Outil | Quand l'utiliser |
|-------|------------------|
| **GitHub Actions** | Repo sur GitHub. Pipelines en YAML. Le plus utilisé aujourd'hui. |
| **Azure Pipelines** (Azure DevOps) | Entreprises avec licences Microsoft, repos privés. |
| **Azure DevOps Boards** | Tickets / sprints, alternative à Jira. |
| **Azure DevOps Artifacts** | Stocker des packages internes (npm, NuGet, PyPI). |

## 7. Git en équipe — branches, PR et stratégies

**Git** est le système de versionnage. En équipe, on ne pousse jamais directement sur `main` : on crée une **branche**, on fait une **Pull Request** (PR), un collègue **relit**, la CI valide, puis on **merge**.

### Anatomie d'une PR

```
 1. Dev crée branche       2. Code + commits       3. Ouvre une PR
    feature/X                  → push                  vers main
                                                          │
                                                          ▼
                                              ┌────────────────────────┐
                                              │ ✅ CI : build + tests   │
                                              │ ✅ Code review (humain) │
                                              │ ✅ Pas de conflit       │
                                              └────────────┬───────────┘
                                                            │ "Approved"
                                                            ▼
                                                      [ merge sur main ]
                                                            │
                                                            ▼
                                                  [ déploiement auto ]
```

### Deux stratégies de branches

| Stratégie | Quand l'utiliser | Branches |
|-----------|------------------|----------|
| **GitFlow** | Releases versionnées (1.0, 1.1, 2.0…). | `main`, `develop`, `feature/*`, `release/*`, `hotfix/*` |
| **Trunk-based** ⭐ | Web / SaaS qui déploie 10×/jour. | `main` + branches courtes `feature/*` (< 2 j) |

**Trunk-based** est devenu la norme dans le cloud — on déploie en continu, pas par grosse release.

### Vocabulaire Git utile

| Terme | Signification |
|-------|---------------|
| `clone` | Télécharger un repo distant. |
| `commit` | Sauvegarder un lot de changements localement. |
| `push` | Envoyer ses commits vers le repo distant. |
| `pull` | Récupérer les derniers changements distants. |
| `merge` | Fusionner deux branches. |
| `rebase` | Réécrire l'historique de sa branche par-dessus une autre (plus propre que merge mais à manier). |
| `conflict` | Deux personnes ont modifié la même ligne → il faut choisir. |
| `tag` | Marquer un commit avec une version (ex. `v1.2.0`). |

## 8. 🎬 Mise en situation 2 — Le cycle de vie d'une Pull Request

**Contexte :** Léo (stagiaire) doit ajouter une route `/health` à l'API. Voici les étapes — exactement ce qui se passe dans 90 % des équipes Cloud.

**1. Récupérer la dernière version**
```bash
git checkout main
git pull
```

**2. Créer une branche**
```bash
git checkout -b feature/health-route
```

**3. Coder + commits atomiques**
```bash
# édite le code...
git add app/routes/health.py
git commit -m "feat(api): add /health route returning 200 OK"
```

**4. Pousser la branche**
```bash
git push -u origin feature/health-route
```

**5. Ouvrir une PR sur GitHub** (bouton ou via `gh pr create`).
- Titre clair : *« feat(api): add /health route »*
- Description : ce qu'on fait, pourquoi, comment tester.
- Lier la story Jira/Azure Boards.

**6. La CI tourne** ✅ build + lint + tests passent.

**7. Code review** par un coéquipier 👀
- *« Pourquoi pas un timeout sur l'appel DB ? »*
- Léo corrige, push à nouveau → la PR s'actualise.

**8. Approbation + merge** 🎉  Le merge déclenche le pipeline CD → la nouvelle route est en **staging** en 5 min, puis en **prod** après approbation manuelle.

**9. Nettoyage**
```bash
git checkout main
git pull
git branch -d feature/health-route
```

> 🧠 **Bonnes pratiques de PR** : petite (idéalement < 400 lignes modifiées), un sujet à la fois, tests inclus, description explicite. Une PR de 3 000 lignes est un cauchemar à relire.

## 9. Et la prod ? — SRE & observabilité

Le code en prod **tombe en panne**. Le rôle qui gère ça : **SRE** (*Site Reliability Engineer*), entre Dev et Ops.

### Notions à connaître

| Terme | Définition |
|-------|------------|
| **SLI** *Service Level Indicator* | Une métrique mesurée (ex. taux d'erreur HTTP). |
| **SLO** *Service Level Objective* | Un objectif interne sur un SLI (ex. < 1 % d'erreurs sur 30 j). |
| **SLA** *Service Level Agreement* | Un engagement contractuel envers les clients (ex. 99,9 % uptime). |
| **Error budget** | La marge entre 100 % et le SLO. Si on la dépasse, on gèle les features et on stabilise. |
| **MTTR** *Mean Time To Recover* | Combien de temps pour réparer en moyenne. |
| **Post-mortem** | Document écrit après un incident : timeline, causes, actions correctives. **Sans blâme.** |

### Astreintes (*on-call*)

Certaines équipes ont une rotation d'astreintes : pendant 1 semaine, vous êtes le contact en cas d'incident prod (alerté par PagerDuty / Teams). C'est la contrepartie du *« vous codez, vous opérez »*.

> 📈 La pratique concrète de l'observabilité (logs, traces, metrics) est dans le **notebook 06**.

## 10. Collaboration au quotidien

### Les outils types

| Type d'outil | Exemples | À quoi ça sert |
|--------------|----------|----------------|
| **Code & PR** | GitHub, Azure Repos, GitLab | Stocker code + faire les revues |
| **Tickets / Backlog** | Azure Boards, Jira, Linear | Suivre user stories & bugs |
| **Documentation** | Confluence, Notion, repo wiki | Décisions, specs, runbooks |
| **Chat** | Microsoft Teams, Slack | Discussions synchrones |
| **Visio** | Teams, Zoom, Google Meet | Réunions à distance |
| **Design** | Figma, Excalidraw, draw.io | Maquettes + schémas |
| **Calendrier partagé** | Outlook, Google Calendar | Rituels + dispos |

### Bonnes pratiques de communication

- **Écrire > parler** pour les décisions importantes (ça reste, c'est partagé, c'est asynchrone).
- **RFC / ADR** *(Request for Comments / Architecture Decision Record)* : un petit doc partagé pour décider d'un changement d'archi. Format type : *Contexte → Options → Décision → Conséquences*.
- **Daily** : 15 min, debout, on tient l'horaire (sinon ça déborde).
- **Async par défaut** : pas besoin de réunion si un message Teams suffit.
- **Demo > slides** : montrer le produit qui tourne vaut 10 PowerPoint.

## 11. 🎬 Mise en situation 3 — Un projet IA agentique de A à Z

**Sujet :** votre équipe doit livrer un **chatbot RH** pour répondre aux questions des employés sur les congés. 3 mois de projet.

### Phase 0 — Cadrage (semaine 1)
- 📋 Atelier avec les RH : quelles questions, quel ton, quels documents source ?
- ✍️ Le PO rédige 5 **user stories** + critères d'acceptation.
- 🎨 L'archi tech part sur : agent Foundry + RAG sur les PDF RH + frontend Teams.
- 📝 RFC validé en revue par le lead.

### Sprint 1 (semaines 2-3) — *MVP technique*
- 🛠️ Provisionne le RG Azure + Foundry (= notebooks 01, 02 du parcours).
- 📥 Ingest des 30 PDF RH dans un index Foundry.
- 🤖 Premier agent avec `file_search`.
- ✅ Démo : « pose-lui une question, il répond avec citation ». Stakeholders content·es.

### Sprint 2 (semaines 4-5) — *Frontend & auth*
- 💬 Intégration dans Teams avec un bot Microsoft Bot Framework.
- 🔐 Authentification Entra ID (les RH ne sont pas tous les employés). Notebook 05.
- 🧪 Tests d'intégration.

### Sprint 3 (semaines 6-7) — *Qualité & sécu*
- 📏 Évaluations Foundry sur 50 questions étalon → 82 % de bonnes réponses.
- 🛡️ Activation du Content Safety + tests des prompts adversariaux.
- 📊 Branche Application Insights (= notebook 06).

### Sprint 4 (semaines 8-9) — *Bêta restreinte*
- 🚀 Déploiement sur 20 utilisateurs pilotes.
- 📝 Collecte feedback dans un canal Teams dédié.
- 🐛 Corrections de bugs, ajustement du system prompt.

### Sprint 5-6 (semaines 10-12) — *Go-live*
- 🌍 Ouvert à tous les employés.
- 📈 Monitoring : taux d'usage, p95 latence, score d'évaluation hebdo.
- 📞 Mise en place d'une rotation d'astreinte (1 personne par semaine).
- 📝 **Post-mortem** d'un incident le jour 3 (« timeout 5xx pendant 12 min ») → actions correctives.

### Run continu
- ♻️ Re-évaluation mensuelle des modèles (drift).
- 📚 Mise à jour de l'index quand le service RH publie un nouveau document.
- 🎯 Roadmap : extension à d'autres domaines (IT, finance…).

> 🧩 Vous voyez tout est lié : les concepts agile + DevOps + observabilité + sécurité que vous venez de lire **se vivent ensemble** dans un projet réel.
